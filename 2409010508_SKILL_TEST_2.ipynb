```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create date range
dates = pd.date_range(start='2025-01-01', periods=365, freq='D')

# Step 2: Generate sample temperature data
np.random.seed(1)
temperature = 25 + 8 * np.sin(np.linspace(0, 2*np.pi, 365)) + np.random.normal(0, 2, 365)

# Step 3: Create DataFrame
data = pd.DataFrame({'Date': dates, 'Temp': temperature})
data.set_index('Date', inplace=True)

# Step 4: Statistical analysis
print("Mean Temperature:", np.mean(data['Temp']))
print("Max Temperature:", np.max(data['Temp']))
print("Min Temperature:", np.min(data['Temp']))

# Step 5: Monthly average
monthly_avg = data.resample('ME').mean()

# Step 6: Rolling average (7 days)
data['Rolling'] = data['Temp'].rolling(window=7).mean()

# Step 7: Plotting
plt.figure(figsize=(10,5))
plt.plot(data['Temp'], label='Daily Temp')
plt.plot(data['Rolling'], label='7-day Avg', color='red')
plt.title("Temperature Time Series")
plt.legend()
plt.grid()
plt.show()

# Monthly plot
monthly_avg.plot(title="Monthly Average Temperature")
plt.show()
```
